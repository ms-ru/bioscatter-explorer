```python
### Imports ###

import requests 
import pandas as pd 
import numpy as np
import plotly.express as px 
```

# Quick content and quality Check


```python
### Data Fetching + quick check ###
url = "https://api.gbif.org/v1/occurrence/search?taxonKey=216&limit=500"
response = requests.get(url)

print(response.status_code)
print(response.content)
```

```python
### Checking Key structure ###

data = response.json()
print(data.keys())

print(data["count"])
print(data["results"])
print(data["facets"])
print(data["endOfRecords"])
print(data["limit"])
print(data["offset"])

```

```python
### Converting to pd-frame ###
df = pd.DataFrame(data["results"])
print(df.columns.to_list)
print(df.shape)
```

```python
### Defining working df ###

# defining needed cols
needed_cols = [
    'order',
    'speciesKey',
    'acceptedTaxonKey',
    'species', 
    'acceptedScientificName',
    'decimalLatitude',
    'decimalLongitude',
    'continent',
    'countryCode',
    'country',
    'stateProvince',
    'locality',
    'year', 
    'month',
    'day',
    'eventDate',
    'individualCount',
    'occurrenceStatus',
    'elevation', 
    'elevationAccuracy',
]

df = df[needed_cols].copy()

# Check
print(df.head())
print(df.shape)
```

```python
### Checking data quality ###

# dict ( key = col index / value 1 = full row count / value 2 = NaN and Null count)
quality_dict = {}

# full frame length
full_length = len(df)

for col in df.columns:
    
    # empty count
    empty = df[col].isna().sum()
    
    # quality_percantage
    percentage = (empty / full_length) * 100
    
    quality_dict[col] = [full_length, empty, percentage]

# Show dict as table    
df_quality = pd.DataFrame.from_dict(quality_dict, orient='index', columns=['full_count', 'missing', 'percent_missing'])

# Sorted
print(df_quality.sort_values(by='percent_missing', ascending=False))
```

# Country Check


```python
### Checking for data counts in different countrys ###

# Fetching data + qick check
country_url = "https://api.gbif.org/v1/occurrence/search?taxonKey=216&limit=0&facet=country&facetLimit=300"
country_response = requests.get(country_url)

print(country_response.status_code)
print(country_response.content)

### Checking Key structure ###
country_data = country_response.json()
print(country_data.keys())

print(country_data["count"])
print(country_data["results"])
print(country_data["facets"])
print(country_data["endOfRecords"])
print(country_data["limit"])
print(country_data["offset"])
```

```python
# converting to data-frame
df_countries = pd.DataFrame(country_data['facets'][0]['counts'])

print(df_countries.columns.to_list)
print(df_countries.shape)
print(df_countries.sort_values(by='count', ascending=False).to_string())
```

# Quick order check for Germany


```python

# Fetching data
url_de = "https://api.gbif.org/v1/occurrence/search?taxonKey=216&country=DE&limit=300"
response_de = requests.get(url_de)

data_de = response_de.json()
print(data_de['count'])
df_de = pd.DataFrame(data_de['results'])
print(df_de['order'].value_counts())

# List with potential candidates
orders = ['Lepidoptera', 'Diptera', 'Hymenoptera', 'Coleoptera', 'Odonata', 'Hemiptera']

for order_name in orders:
    # get taxon key for order 
    match_url = f"https://api.gbif.org/v1/species/match?name={order_name}&rank=ORDER"
    match = requests.get(match_url).json()
    taxon_key = match['usageKey']
    
    # counting listings for DE
    count_url = f"https://api.gbif.org/v1/occurrence/search?taxonKey={taxon_key}&country=DE&limit=0"
    count = requests.get(count_url).json()['count']
    print(f"{order_name} (key={taxon_key}): {count:,}")
```
